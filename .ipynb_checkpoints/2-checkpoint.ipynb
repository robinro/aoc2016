```python
keypad = {
    1j-1: 1, 1j: 2, 1j+1: 3,
    -1: 4, 0: 5, 1: 6,    
    -1j-1: 7, -1j: 8, -1j+1: 9
}
directions = {
    'U': 1j,
    'D': -1j,
    'R': +1,
    'L': -1,
}
reversekeypad = {v: k for k,v in keypad.items()}
```

```python
def limit(pos):
    return max(-1, min(1, pos.real)) + 1j*max(-1, min(1, pos.imag))
```

```python
def getkey(moves, start=5):
    pos = reversekeypad[start]
    for move in moves:
        newpos =pos+directions[move]
        pos = limit(newpos)
    return keypad[pos]
```

```python
getkey('RRDDD')
```

```python
def lines(ll, start=5, getkey=getkey):
    pos = start
    for l in ll.splitlines():
        #print(l)
        pos = getkey(l, pos)
        print(pos)
```

```python
keypad2 = {
    2j: 1, 
    1j-1: 2, 1j:3, 1j+1:4,
    -2:5, -1:6, 0:7, 1:8, 2:9,
    -1j-1: 'A', -1j: 'B', -1j+1: 'D',
    -2j: 'D',
    }
reversekeypad2 = {v:k for k,v in keypad2.items()}
```

```python
def step_keypad(moves, start=5, keypad=keypad2, reversekeypad=reversekeypad2):
    pos = reversekeypad[start]
    for move in moves:
        newpos =pos+directions[move]
        if newpos in keypad:
            pos = newpos
    return keypad[pos]
```

```python
lines('''ULL
RRDDD
LURDL
UUUUD''', getkey=step_keypad)
```
